```python
import pandas as pd
```

```python
# load pairwise dataset
fn = '/fs/nexus-scratch/rhaworth/plasmid/pairs.txt'
data = pd.read_csv(fn, sep='\t')
data
```

```python
# load 4mers
fn = '/fs/nexus-projects/plasmids/plasmid_host_range_2024_summer_interns/expanded_dataset/4mer_table.txt'
data_4mers = pd.read_csv(fn, sep='\t')
data_4mers
```

```python
# split 4mers into chromosomes/plasmids
c_idx = data_4mers.index[data_4mers['Seq_type'] == 'chromosome']
c_4mers = data_4mers.iloc[c_idx]
p_4mers = data_4mers.iloc[~c_idx]
print(c_4mers['Seq_type'].unique())
print(p_4mers['Seq_type'].unique())
```

```python
# get chromosome 4mers in order of pairwise dataset
labels = data['Assembly_ID'].str[:-7] # strip filename
_c4 = c_4mers.set_index('SeqID')
# a few labels are not found in c_4mers, so just drop them i guess
labels_to_keep = labels.isin(c_4mers['SeqID'])
labels = labels[labels_to_keep]

c4m = _c4.loc[labels]
c4m
```

```python
labels = data['Plasmid_ID']
# drop labels
labels = labels[labels_to_keep]
_p4 = p_4mers.set_index('SeqID')

labels_to_keep_p = labels.isin(p_4mers['SeqID'])
labels = labels[labels_to_keep_p]

p4m = _p4.loc[labels]
p4m
```

```python
# convert to kmer frequencies
# ignore terrible variable naming conventions idk what i'm doing here
data_c = data[labels_to_keep]
data_p = data_c[labels_to_keep_p]
p4m_swap = p4m.copy()
p4m_swap['SeqID'] = p4m.index
p4m_swap['Index'] = data_p.index
p4m_swap = p4m_swap.set_index('Index')
p_kmers = p4m_swap.iloc[:,2:-1]
p_sums = p_kmers.sum(axis=1)
p_freqs = p_kmers.div(p_sums, axis=0)

c4m['Index'] = data_c.index
c4m_red = c4m.set_index('Index')[labels_to_keep_p]
c_kmers = c4m_red.iloc[:,2:]
c_sums = c_kmers.sum(axis=1)
c_freqs = c_kmers.div(c_sums, axis=0)

# compute differences
diffs = p_freqs.add(c_freqs * -1)
diffs
```

```python
# convert to RMSD across all kmers
rmsd = diffs ** 2
rmsd = rmsd.sum(axis=1)
rmsd = rmsd / 136 # * 1/n
rmsd = rmsd ** 0.5
# mean + std of RMSD across all samples
print(rmsd.mean(), '+/-', rmsd.std())
```

```python
# add full kmer profiles or just RMSD
full = False
if full:
    data_ext = pd.concat([data_p, diffs], axis=1)
else:
    data_ext = data_p.copy()
    data_ext['RMSD'] = rmsd
data_ext
```

```python
# compute other pairwise features
data_ext['Len_ratio'] = data_ext['Chr_len'].div(data_ext['Plasmid_len'])
data_ext['GC_abs_diff'] = data_ext['Chr_GC'].add(data_ext['Plasmid_GC'] * -1).abs()
data_ext
```

```python
# fetch SRCCs; takes a few minutes due to large file size
data_ext['SRCC'] = 0.0

# first trim assembly IDs
assemblies_trimmed = data_ext['Assembly_ID'].str[:-7].values

fn = '/fs/nexus-projects/plasmids/plasmid_host_range_2024_summer_interns/expanded_dataset/correlation_table_v4.txt'
index = 0
populated = 0
with open(fn, 'r') as f:
    # save list of plasmids
    plasmid_list = f.readline().split()
    while True:
        if index < 6800: # found in practice the first 6800 lines aren't present in data_ext, so this makes things faster
            next(f)
            index += 1
            continue
        line = f.readline()
        if line is None or line == '':
            break
        assembly = line.split('\t', 1)[0]
        assembly = assembly[:-15]
        # check for containment; assume all assemblies found
        if assembly in assemblies_trimmed:
            rows_found = data_ext[data_ext['Assembly_ID'].str[:-7] == assembly]
            for i, plasmid in enumerate(rows_found['Plasmid_ID']):
                if plasmid not in plasmid_list:
                    # drop row if plasmid not found
                    data_ext.drop(rows_found.iloc[i].name, inplace=True)
                else:
                    # if both assembly and plasmid are found, split line further to extract plasmid
                    plasmid_idx = plasmid_list.index(plasmid)
                    # if idx=1, we need 3 splits: 1 for assembly ID, 2 for ID + idx 0, 3 for ID + idx 0-1
                    # idx+1 indexing works in last index edge case, -2 indexing doesn't
                    srcc = line.split('\t', plasmid_idx+2)[plasmid_idx+1]
                    # update dataframe
                    data_ext.loc[rows_found.iloc[i].name, 'SRCC'] = float(srcc)
                    populated += 1
        index += 1
        if index % 5000 == 0:
            print('iter', index)
print(populated, 'rows populated out of', len(data_ext))
data_ext
```
